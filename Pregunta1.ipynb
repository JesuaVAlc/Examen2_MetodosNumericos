```python
from scipy.optimize import bisect, newton
import numpy as np
import matplotlib.pyplot as plt
```

```python
ecuacion = lambda x: x**5 + 5.5*x**4 + 4.5*x**3 - 9.5*x**2 -15.5*x -6
#grafica para ver la funcion
x = np.linspace(-5, 2, 100)
plt.plot(x, ecuacion(x), label='f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Gráfica de la función')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(-5, 2)
plt.ylim(-10, 10)
plt.grid()
plt.legend()
plt.show()
```
