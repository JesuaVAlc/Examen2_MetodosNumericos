```python
from scipy.optimize import bisect, newton
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**5 + 5.5*x**4 + 4.5*x**3 - 9.5*x**2 - 15.5*x - 6
#grafica para ver la funcion
x = np.linspace(-7, 3, 100)
y = f(x)

plt.figure(figsize=(10,6))
plt.plot(x, y, label="f(x)")
plt.axhline(0, color='black', linestyle='--')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.xlim(-1.2, -0.08)
plt.ylim(-10, 10)
plt.title("Gráfica de f(x) para estimar intervalos de raíces")
plt.grid(True)
plt.legend()
plt.show()
```

```python
intervalos = [(-7, -6), (-6, -5), (-4, -3), (-2, -1), (0, 1), (1, 2)]

raices_biseccion = []
for a, b in intervalos:
    try:
        raiz = bisect(f, a, b)
        raices_biseccion.append(raiz)
    except ValueError:
        pass

print("Raíces encontradas por bisección:")
for r in raices_biseccion:
    print(f"{r:.6f}")
```

```python
raices_newton = []
for a, b in intervalos:
    try:
        raiz = newton(f, (a + b) / 2)
        raices_newton.append(raiz)
    except RuntimeError:
        pass    
print("Raíces encontradas por Newton:")
for r in raices_newton:
    print(f"{r:.6f}")
```
